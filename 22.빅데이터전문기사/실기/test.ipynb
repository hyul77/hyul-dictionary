```python
train_y = train['survived']
train_x = train.copy()
train_x.drop(columns=['survived'], inplace=True)
```

```python
train_x.info()
```

```python
train_x_num = train_x.select_dtypes(include=np.number)
train_x_num.info()
```

```python
train_x_cat = train_x.select_dtypes(exclude=np.number)
train_x_cat.info()
```

```python
train_x.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

for col in train_x_cat.columns:
    train_x_cat[col] = encoder.fit_transform(train_x_cat[col].astype(str))
    
train_x[train_x_cat.columns] = train_x_cat
```

```python
train_x.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()

test_cat = test.select_dtypes(exclude=np.number)

for col in test_cat.columns:
    test_cat[col] = encoder.fit_transform(test_cat[col].astype(str))

test[test_cat.columns] = test_cat
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


model = RandomForestClassifier(random_state=42)
model.fit(train_x, train_y)

pred = model.predict(test)
print('예측 정확도 : ', accuracy_score(valid, pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor(random_state=42)
model2.fit(train_x, train_y)
```

```python
pred = model2.predict(test)
print('예측 정확도 : ', accuracy_score(valid, pred))
```
